```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
xs = np.arange(-1, 5, dtype=float)
print(xs)
ys = np.array([-4.0, 1.0, 6.0, 11.0, 16.0, 21.0], dtype=float)
print(ys)
# print(f'xs = {xs}, ys = {ys}')
```

```python
plt.plot(xs, ys, 'ro')
plt.show()
```

```python
# QuГ© modelo voy a utilizar
regresion_lineal = np.polyfit(xs, ys, 1)

beta, alfa = regresion_lineal

print(f'La ecuaciГіn de la recta es: y = {round(beta, 1)}x + {round(alfa, 1)}')
```

```python
plt.plot(xs, ys, 'o', color='green', label='Datos')
plt.plot(xs, beta*xs + alfa, color='blue', label='Ajuste a la RegresiГіn Lineal')
plt.axhline(y = 0, color = 'orange')
plt.axvline(x = 0, color = 'orange')
plt.title('RegresiГіn Lineal')
plt.xlabel('TamaГ±o de la casa')
plt.ylabel('Precio de la casa')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PerceptrГіn simple
capa = tf.keras.layers.Dense(input_shape=[1], units=1)
modelo = tf.keras.Sequential([capa])
modelo.compile(
  optimizer="sgd",
  loss="mean_squared_error"
)
```

```python
print('Hola, empezarГ© a entrenar...')
hist = modelo.fit(xs, ys, epochs=500, verbose=True)
modelo, hist
```

```python
valor_a_predicir = float(input('Ingresa cuantas habitaciones tiene la casa para conocer su precio: '))
prediccion = modelo.predict(np.array([[valor_a_predicir]]))
print(f'El precio de la casa con {valor_a_predicir} habitaciones es: {prediccion}')
```

Visualice los datos a travГ©s de un grГЎfico para analizar el comportamiento de una regresiГіn lineal con los datos en **Celsius** y **Fahrenheit**.

$Recordando$ $la$ $ecuaciГіn$ $para$ $la$ $conversiГіn$ $de$ $grados$ $Celcius$ $a$ $Fahrenheit:$

$
FВ° = CВ°*1.8 + 32
$

Entrene una red neuronal para predecir los datos siguientes, recordando que, la infraestructura de prueba, espera un modelo entrenado que acepte una forma de entrada de [1].

Realice un test de entrenamiento para verificar las epocas dadas y visualice el comportamiento de las perdidas segГєn el algoritmo utilizado.  

Guarde el modelo anterior en un archivo con el nombre y tipo de dato **mymodel.h5**

```python
# Datos Celsius -> Fahrenheit
celsius = np.array([-40.0, -10.0, 0.0, 8.0, 15.0, 22.0, 38.0], dtype=float)
fahrenheit = np.array([-40.0, 14.0, 32.0, 46.0, 59.0, 72.0, 100.0], dtype=float)
print('Celsius:', celsius)
print('Fahrenheit:', fahrenheit)
```

```python
# Visualizar y ajuste lineal
plt.figure(figsize=(6,4))
plt.scatter(celsius, fahrenheit, color='red', label='Datos')
m, b = np.polyfit(celsius, fahrenheit, 1)
plt.plot(celsius, m*celsius + b, color='blue', label=f'Ajuste: y={m:.2f}x+{b:.2f}')
plt.title('Celsius vs Fahrenheit')
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Entrenar una red neuronal simple
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='adam', loss='mean_squared_error')
print('Entrenando...')
history = model.fit(celsius, fahrenheit, epochs=500, verbose=0)
print('Entrenamiento completado')
# Mostrar la pГ©rdida
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'])
plt.title('Loss durante el entrenamiento')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
# Predicciones y comparaciГіn
preds = model.predict(celsius).flatten()
for c, f_true, f_pred in zip(celsius, fahrenheit, preds):
    print(f'{c:.1f}В°C -> real: {f_true:.1f}В°F, pred: {f_pred:.2f}В°F')
plt.figure(figsize=(6,4))
plt.scatter(celsius, fahrenheit, label='Real')
plt.scatter(celsius, preds, label='PredicciГіn')
plt.plot(celsius, m*celsius + b, color='gray', linestyle='--', label='Ajuste lineal')
plt.legend()
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.grid(True)
plt.show()
# Guardar el modelo en la carpeta output (formato Keras nativo .keras)
import os
os.makedirs('output', exist_ok=True)
model.save('output/mymodel.keras')
print('Modelo guardado en output/mymodel.keras')
```
